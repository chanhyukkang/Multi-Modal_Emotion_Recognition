## Extract Audio Features

```python
# features : signal mean, signal std, rmse mean, rmse std, silence, harmonic, auto correlation max, auto correlation std

import os
import pickle
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as ms
from tqdm import tqdm
import librosa
import math
import random
import pandas as pd
import IPython.display
import librosa.display

df = pd.read_csv('C:/Users/KangChanhyuk/Desktop/자주연1/audio_test_speechhhhh.csv', engine = 'python') 
# df = pd.read_csv('C:\\Users\KangChanhyuk\Desktop\자주연1\IEMOCAP\iemocap.csv', engine = 'python') 
print(len(df))

for i in range(8599, len(df) ) :
    file_path = 'C:\\Users\KangChanhyuk\Desktop\자주연1\Friends\Friends_Audio_Speech\S'+str(df['Season'][i])+'\\'+filename[i]+'\\vocals.wav'
    # file_path = 'C:\\Users\KangChanhyuk\Desktop\자주연1\IEMOCAP\\'+str(df["Ses"][i])+'\\'+str(df["Ses_name"][i])+'.wav'
    
    y, sr = librosa.load(file_path, sr=44100)
    
    df.loc[i, ["signal_mean"]] = np.mean(abs(y))
    df.loc[i, ["signal_std"]] = np.std(y)
    
    rmse = librosa.feature.rms(y + 0.0001)[0]
    df.loc[i, ["rmse_mean"]] = np.mean(rmse)
    df.loc[i, ["rmse_std"]] = np.std(rmse)
    
    silence = 0
    for e in rmse:
        if e <= 0.4 * np.mean(rmse):
            silence += 1
    df.loc[i, ["silence"]] = silence/float(len(rmse))
    
    y_harmonic, y_percussive = librosa.effects.hpss(y)
    df.loc[i, ["harmonic"]] = np.mean(y_harmonic)
    
    autocorr = librosa.core.autocorrelate(y)
    df.loc[i, ["autocorr_max"]] = np.max(autocorr)

    cl = 0.45 * np.mean(abs(y))
    center_clipped = []
    for s in y:
        if s >= cl:
            center_clipped.append(s - cl)
        elif s <= -cl:
            center_clipped.append(s + cl)
        elif np.abs(s) < cl:
            center_clipped.append(0)
    new_autocorr = librosa.core.autocorrelate(np.array(center_clipped))
    df.loc[i, ["autocorr_std"]] = np.max(new_autocorr)
    
    df.to_csv("C:/Users/KangChanhyuk/Desktop/자주연1/audio_test_speechhhhh.csv", mode='w')
    
    print(i)

print("done")
```

```python
 import pandas as pd
import numpy as np
import xgboost as xgb
import pickle
import matplotlib
import matplotlib.pyplot as plt
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.feature_selection import SelectFromModel
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import itertools
from IPython.display import display

x_train = pd.read_csv('C:/Users/KangChanhyuk/Desktop/자주연1/iemocap_emo4.csv')
x_test = pd.read_csv('C:/Users/KangChanhyuk/Desktop/자주연1/audio_test_emo4.csv')

# x_train = pd.read_csv('C:/Users/KangChanhyuk/Desktop/자주연1/iemocap_emotion.csv')
# x_test = pd.read_csv('C:/Users/KangChanhyuk/Desktop/자주연1/audio_test_emotion.csv')

# scalar = MinMaxScaler()
# x_test[x_test.columns[1:]] = scalar.fit_transform(x_test[x_test.columns[1:]])

print(x_train.shape)
print(dict(pd.Series(x_train['emotion']).value_counts()))
print(x_test.shape)
#print(dict(pd.Series(x_test['emotion']).value_counts()))
print(dict(pd.Series(x_test['Emotion_SUMMAX']).value_counts()))
###############################################
# y_train = x_train['Emotion_SUMMAX']
# y_test = x_test['Emotion_SUMMAX']

# cl_weight = dict(pd.Series(x_train['Emotion_SUMMAX']).value_counts(normalize=True))
# print(dict(pd.Series(x_train['Emotion_SUMMAX']).value_counts()))

# del x_train['Emotion_SUMMAX']
# del x_test['Emotion_SUMMAX']

# emotion_dict = {'anger': 0,
#                 'happiness': 1,
#                 'sadness': 2,
#                 'fear': 3,
#                 'surprise': 4,
#                 'neutral': 5,
#                 'disgust': 6,
#                }

# emo_keys = list(['anger', 'happiness', 'sadness', 'fear', 'surprise', 'neutral','disgust'])

##############################################

y_train = x_train['emotion']
y_test = x_test['Emotion_SUMMAX']
cl_weight = dict(pd.Series(x_train['emotion']).value_counts(normalize=True))
print(dict(pd.Series(x_train['emotion']).value_counts()))

# y_train = x_train['sentiment']
# y_test = x_test['OUR_Sentiment']
# cl_weight = dict(pd.Series(x_train['sentiment']).value_counts(normalize=True))
# print(dict(pd.Series(x_train['sentiment']).value_counts()))



# del x_train['Ses']
# del x_train['Ses_name']
del x_train['emotion']
# del x_train['emotion_num']
# del x_train['sentiment']
# del x_train['sentiment_num']

del x_test['Emotion_SUMMAX']
#del x_test['OUR_Sentiment']
# del x_test['MELD_Sentiment']
# del x_test['EmoFreq_Sentiment']
# del x_test['OUR_Sentiment']


#emotion_dict = {'positive': 1, 'negative': 2, 'neutral': 0, }
#emo_keys = list(['neutral','positive','negative'])

#emotion_dict = {'hap':0, 'sad':1, 'neu':2, 'ang':3, 'exc':4, 'fru':5}
#{'positive': 1, 'negative': 2, 'neutral': 0, }
#######################################################
emotion_dict = {'hap':0, 'sad':1, 'neu':2, 'ang':3}
# #emo_keys = list(['hap','sad','neu', 'ang', 'exc', 'fru'])
 emo_keys = list(['hap','sad','neu', 'ang'])
#####################################################
#emotion_dict = {'pos':0, 'pneu':1, 'neu':2, 'nneu':3, 'neg':4}
#emo_keys = list(['hap','sad','neu', 'ang', 'exc', 'fru'])
#emo_keys = list(['pos','pneu','neu', 'nneu', 'neg'])

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    # plt.figure(figsize=(8,8))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
def one_hot_encoder(true_labels, num_records, num_classes):
    temp = np.array(true_labels[:num_records])
    true_labels = np.zeros((num_records, num_classes))
    true_labels[np.arange(num_records), temp] = 1
    return true_labels

def display_results(y_test, pred_probs, cm=True):
    pred = np.argmax(pred_probs, axis=-1)
    print(pred)
    one_hot_true = one_hot_encoder(y_test, len(pred), len(emotion_dict))
    print('Test Set Accuracy =  {0:.3f}'.format(accuracy_score(y_test, pred)))
    print('Test Set F-score =  {0:.3f}'.format(f1_score(y_test, pred, average='macro')))
    print('Test Set Precision =  {0:.3f}'.format(precision_score(y_test, pred, average='macro')))
    print('Test Set Recall =  {0:.3f}'.format(recall_score(y_test, pred, average='macro')))
    if cm:
        plot_confusion_matrix(confusion_matrix(y_test, pred), classes=emo_keys)
        
print(y_train)
```
